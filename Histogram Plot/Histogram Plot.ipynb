```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Hist Plot 

```python
#DataSet Link:-https://www.kaggle.com/datasets/pyim59/basic-datasets?resource=download&select=penguins.csv
```

```python
var=sns.load_dataset('penguins')
var
```

```python
sns.distplot(var['bill_length_mm'],color='violet')
plt.show()
```

```python
sns.distplot(var['flipper_length_mm'],color='m')
plt.show()
```

```python
sns.distplot(var['flipper_length_mm'],bins=[170,180,190,210,220,230,240],color='grey')
plt.show()
```

```python
#Kernel Density extimater

sns.distplot(var['flipper_length_mm'],kde=True,color='black')
plt.show()
```

```python

sns.distplot(var['flipper_length_mm'],kde=True,color='m',rug=True)
plt.show()
```

```python
sns.distplot(var['flipper_length_mm'],kde=True,rug=True,color='r')
plt.show()
```

```python
sns.histplot(var['flipper_length_mm'], kde=True, color='g', log_scale=True)
plt.show()
```
