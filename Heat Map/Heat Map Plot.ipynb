```python
y={'fontsize':20,'color':'g'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=5,linecolor='grey',
            cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
y={'fontsize':20,'color':'g'}

v = sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=5,linecolor='grey',
            cbar=False,xticklabels=False,yticklabels=False)

v.set(xlabel='python',ylabel='java')
plt.show()
```

```python
y={'fontsize':20,'color':'g'}

v = sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=5,linecolor='grey',
            cbar=False,xticklabels=False,yticklabels=False)

v.set(xlabel='python',ylabel='java')
sns.set(font_scale=2)
plt.show()
```
