```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var =  sns.load_dataset('tips')
```

```python
var
```

```python
sns.countplot(x='sex',data=var)
plt.show()
```

```python
sns.countplot(x='sex',data=var,palette='Accent')
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=var,palette='coolwarm')
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=var,palette='coolwarm',hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',data=var,palette='Accent',hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=var,palette='Accent',hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=var,color='r',hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=var,color='r',hue='smoker',saturation=0.1)
plt.show()
```
