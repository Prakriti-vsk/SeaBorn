```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var =  sns.load_dataset('tips')
```

```python
var
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',hue='time',data=var)
plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='time', data=var)

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='g',data=var)

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='g',data=var,order=['Fri','Sun','Thur','Sat'])

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='g',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True)

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill',color='g',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True)

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='g',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'})

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='g',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},linewidth=2)

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='r',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},linewidth=2,palette='BuPu')

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='r',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},linewidth=2,palette='plasma')

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', hue='sex',color='r',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},linewidth=2,palette='BuPu',orient='v')

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='total_bill',y='day', hue='sex',color='r',data=var,order=['Fri','Sun','Thur','Sat'],
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},linewidth=2,palette='BuPu',orient='h')

plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(data=var,
           showmeans=True,meanprops={'marker':'+','markeredgecolor':'g'},
            linewidth=2,palette='BuPu',orient='h')

plt.show()
```
