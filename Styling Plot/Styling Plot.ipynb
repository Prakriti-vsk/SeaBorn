```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.boxplot(y='total_bill',data=var)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill',data=var)

plt.grid()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=var)

plt.grid()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.grid()
```

```python
plt.figure(figsize=(3,12))
sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.grid()
```

```python
plt.figure(figsize=(5,7))
sns.set_style('whitegrid')
sns.set_context('poster',font_scale=1)
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.grid()
```

```python
plt.figure(figsize=(5,7))
sns.set_style('whitegrid')
sns.set_context('paper',font_scale=1)
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.grid()
```

```python
plt.figure(figsize=(5,7))
sns.set_style('whitegrid')
sns.set_context('paper',font_scale=1)
sns.barplot(x='day',y='total_bill',data=var,palette='Accent')
sns.despine()
plt.grid()
```

```python
plt.figure(figsize=(5,7))
sns.set_style('whitegrid')
sns.set_context('paper',font_scale=1)
sns.barplot(x='day',y='total_bill',data=var,palette='cool')
sns.despine()
plt.grid()
```
