```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
fg= sns.FacetGrid(var,col='sex')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.scatter,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='cool')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='winter')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='summer')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='spring')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='autumn')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```

```python
fg= sns.FacetGrid(var,col='sex',hue='day',palette='cool')
fg.map(plt.bar,'total_bill','tip',edgecolor='m').add_legend()

plt.show()
```
