```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips').head(20)
```

```python
var
```

# Cat Plot

```python
sns.catplot(x='tip',y='size',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex')
plt.show()
```

```python
sns.catplot(x='tip',y='total_bill',data=var,hue='sex',palette='Oranges',kind='bar')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',palette='Oranges',kind='bar')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',palette='Oranges',kind='point')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',palette='Oranges',kind='boxen')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',palette='Oranges',kind='violin')
plt.show()
```

```python
sns.set_style('white')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
sns.set_context('paper',font_scale=0.3)
plt.show()
```

```python
sns.set_style('white')
sns.barplot(x='day',y='total_bill',data=var,palette='summer')
sns.despine()
sns.set_context('poster',font_scale=0.3)
plt.show()
```

# Multiple Grid(Facet-Grid)

```python
fg = sns.FacetGrid(var)
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex',hue='day',palette='BuPu')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col='sex',hue='day',palette='BuPu')
fg.map(plt.bar,'total_bill','tip',edgecolor='r').add_legend()
plt.show()
```
