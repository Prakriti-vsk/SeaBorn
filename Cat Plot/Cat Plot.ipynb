```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.catplot(x='tip',y='size',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Accent',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Oranges',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Accent',height=10,data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Accent',kind='bar',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Accent',kind='box',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',hue='sex',palette='Accent',kind='boxen',data=var)
plt.show()
```
