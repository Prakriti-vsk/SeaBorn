```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var= sns.load_dataset('tips')
```

```python
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue='sex')
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex')
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='BuPu')
plt.show()
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='BuGn')
plt.show()
```

```python
sns.pairplot(var,x_vars=['total_bill','tip'],y_vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='BuPu')
plt.show()
```

```python
sns.pairplot(var,x_vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='BuPu')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='reg')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='scatter')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='hist')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='kde')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='reg',diag_kind='kde')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuPu',kind='kde',diag_kind='kde')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuGn',markers=['*','<'])
plt.show()
```
