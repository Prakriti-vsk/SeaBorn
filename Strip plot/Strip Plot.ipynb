```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Strip Plot

```python
var =  sns.load_dataset('tips')
var
```

```python
sns.stripplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,color='r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r',jitter=12)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r',jitter=12,size=20)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r',jitter=12,size=20,marker='o')
plt.show()
```

```python
m={'Male':'*','Female':'o'}
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r',jitter=12,size=20,marker='<')
plt.show()
```

```python
m={'Male':'*','Female':'o'}
sns.stripplot(x='day',y='total_bill',hue='sex',data=var,palette='BuPu',linewidth=2,edgecolor='r',jitter=12,size=20,marker='<',alpha=0.2)
plt.show()
```
