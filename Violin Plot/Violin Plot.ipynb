```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.violinplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='time',data=var)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2,color='White')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2,palette='BuPu')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2,palette='Dark2')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner','Lunch'],palette='Accent')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner','Lunch'],palette='Accent',saturation=0.5)
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner','Lunch'],color='y')
plt.show()
```
