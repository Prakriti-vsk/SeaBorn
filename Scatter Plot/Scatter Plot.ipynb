```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
#DataSet Link:-https://www.kaggle.com/datasets/pyim59/basic-datasets?resource=download&select=penguins.csv
```

```python
var=sns.load_dataset('penguins').head(20)
var
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(78,56))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(78,56),palette='Accent')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(78,56),palette='gist_ncar',alpha=0.56)
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(78,56),
                palette='gist_ncar',alpha=0.56,markers=m)
plt.show()
```
