```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Bar Plot

```python
#DataSet Link:-https://www.kaggle.com/datasets/pyim59/basic-datasets?resource=download&select=penguins.csv
```

```python
var=sns.load_dataset('penguins')
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=var,color='r')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=var,hue='sex')
plt.show()
```

```python
order_1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=var,hue='sex',order=order_1)
plt.show()
```
